```python
'''ASSIGNMENT 2

TITLE: McCulloch-Pitts neural Network

PROBLEM STATEMENT: -
Generate ANDNOT function using McCulloch-Pitts neural net by a python program.'''

# Define the activation function (step function)
def activation(weighted_sum, threshold):
    return 1 if weighted_sum >= threshold else 0

# Inputs for AND-NOT gate (X1 AND NOT X2)
inputs = [
    (0, 0),
    (0, 1),
    (1, 0),
    (1, 1)
]

# Define weights and threshold
# For AND-NOT: Output = X1 AND (NOT X2)
# So use: weight1 = 1, weight2 = -1, threshold = 1
w1 = 1
w2 = -1
threshold = 1

# Display results
print("X1 X2 | Output (X1 AND NOT X2)")
print("-----------------------------")
for x1, x2 in inputs:
    weighted_sum = w1 * x1 + w2 * x2
    output = activation(weighted_sum, threshold)
    print(f" {x1}  {x2}  |        {output}")

'''Here are the answers to your questions about the **McCulloch-Pitts (MCP) Model**, also known as the **MCP Neuron**:

---

### **Q1. Explain MCP Model?**

The **McCulloch-Pitts (MCP) Model** is one of the earliest mathematical models of a biological neuron, proposed in 1943. It is a simplified representation of how a neuron processes binary input to produce binary output.

**Key Components:**

* **Inputs (x₁, x₂, ..., xₙ):** Binary values (0 or 1)
* **Weights (w₁, w₂, ..., wₙ):** Represent the importance of each input (usually 1 in MCP)
* **Summation Function:** Calculates the weighted sum of the inputs

  $$
  \text{Net Input} = \sum_{i=1}^n w_i \cdot x_i
  $$
* **Threshold (θ):** If the net input ≥ θ, neuron outputs 1; else 0
* **Activation Function:** A step function (binary output)

**Output Function:**

$$
Y = 
\begin{cases}
1 & \text{if } \sum w_i x_i \geq \theta \\
0 & \text{otherwise}
\end{cases}
$$

---

### **Q2. How to generate AND NOT function using MCP Model?**

To implement the **AND NOT** logic:
**Function:** $Y = A \cdot \overline{B}$

| A | B | Output |
| - | - | ------ |
| 0 | 0 | 0      |
| 0 | 1 | 0      |
| 1 | 0 | 1      |
| 1 | 1 | 0      |

**Step-by-step MCP setup:**

* Inputs: A, B
* Weights: Let w₁ = 1 (for A), w₂ = -1 (for NOT B)
* Threshold: θ = 1

**Net Input:**

$$
Y = 
\begin{cases}
1 & \text{if } (1 \cdot A) + (-1 \cdot B) \geq 1 \\
0 & \text{otherwise}
\end{cases}
$$

**This gives:**

* (1, 0): 1 → ✅
* All other combinations: < 1 → 0

This satisfies AND NOT logic.

---

### **Q3. Give McCulloch-Pitts Artificial Neuron Model Limitations**

1. **Binary Outputs Only:** Can only produce 0 or 1 as output.
2. **No Learning:** MCP doesn't learn from data; weights and threshold are fixed.
3. **No Handling of Noise:** Inputs must be clean and binary; can't deal with real-world data well.
4. **Only Linearly Separable Problems:** Cannot solve problems like XOR.
5. **No Weight Adaptation:** Unlike modern neurons, MCP doesn't use learning algorithms like gradient descent.
6. **Fixed Weights:** Weights are manually assigned, not trained.
7. **No Continuous Activation Functions:** Can't model non-binary or probabilistic behaviors.

---

Let me know if you'd like a diagram or Python code to demonstrate the MCP model.
'''
```
