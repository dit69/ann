```python
""" 
Group A: Assignment No. 1
Assignment Title: Write a Python program to plot a few activation
functions that are being used in neural networks.
"""
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    exp_values = np.exp(x - np.max(x))  # for numerical stability
    return exp_values / np.sum(exp_values, axis=0, keepdims=True)

# Linear activation function
def linear(x):
    return x

# Input values
x_values = np.linspace(-5, 5, 100)

# Plotting each activation function
plt.figure(figsize=(15, 10))

plt.subplot(2, 4, 1)
plt.title('Sigmoid Function')
plt.plot(x_values, sigmoid(x_values))
plt.grid(True)

plt.subplot(2, 4, 2)
plt.title('Hyperbolic Tangent Function (tanh)')
plt.plot(x_values, tanh(x_values))
plt.grid(True)

plt.subplot(2, 4, 3)
plt.title('ReLU (Rectified Linear Unit)')
plt.plot(x_values, relu(x_values))
plt.grid(True)

plt.subplot(2, 4, 4)
plt.title('Softmax')
input_softmax = np.array([2.0, 1.0, 0.1])
plt.bar(range(len(input_softmax)), softmax(input_softmax))
plt.grid(True)

plt.subplot(2, 4, 5)
plt.title('Linear')
plt.plot(x_values, linear(x_values))
plt.grid(True)

plt.tight_layout()
plt.show()

'''Here are the detailed answers for your assignment on **Activation Functions in Neural Networks**:

---

### **1) Explain the role of activation functions in artificial neural networks.**

Activation functions introduce **non-linearity** into the neural network, enabling it to learn complex patterns. Without activation functions, the network would behave like a linear regression model, limiting its ability to solve problems like image recognition or natural language processing. They help neurons decide whether to "fire" or remain inactive based on the input signal.

---

### **2) Compare and contrast linear and nonlinear activation functions. Provide examples of each.**

| Aspect       | Linear Activation Function      | Nonlinear Activation Function     |
| ------------ | ------------------------------- | --------------------------------- |
| Nature       | Straight line                   | Curve or step-like                |
| Output Range | Unbounded (- to )             | Often bounded (e.g., 01 or -11) |
| Usefulness   | Limited to simple relationships | Captures complex patterns         |
| Gradient     | Constant                        | Varies with input                 |
| Examples     | $f(x) = x$                      | Sigmoid, tanh, ReLU, Leaky ReLU   |

---

### **3) Define the sigmoid and hyperbolic tangent (tanh) activation functions.**

* **Sigmoid:**

  $$
  \sigma(x) = \frac{1}{1 + e^{-x}}  
  $$

  Output: (0, 1)

* **Tanh:**

  $$
  \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}  
  $$

  Output: (-1, 1)

---

### **4) Discuss the advantages and disadvantages of using sigmoid and tanh functions in neural networks.**

| Function | Advantages                             | Disadvantages                                     |
| -------- | -------------------------------------- | ------------------------------------------------- |
| Sigmoid  | Good for binary classification; smooth | Saturates at extremes; vanishing gradient problem |
| Tanh     | Zero-centered; better than sigmoid     | Still suffers from vanishing gradients            |

---

### **5) Describe the Rectified Linear Unit (ReLU) activation function.**

* **Definition:**

  $$
  f(x) = \max(0, x)
  $$

* **Advantages:**

  * Avoids vanishing gradient in most cases
  * Fast convergence
  * Sparse activation

* **Disadvantages:**

  * Can "die" if too many neurons output 0 (Dead ReLU problem)

---

### **6) Why is the sigmoid activation function commonly used in the output layer of binary classification problems?**

Sigmoid outputs a value between 0 and 1, which can be interpreted as a **probability**. This makes it ideal for binary classification where the output needs to represent the likelihood of a class.

---

### **7) Provide examples of scenarios where one activation function might be more suitable than another.**

* **Sigmoid:** Binary classification (e.g., spam vs not spam)
* **Tanh:** When inputs need to be normalized between -1 and 1
* **ReLU:** Hidden layers in image classification (e.g., CNNs)
* **Softmax:** Multi-class classification in the output layer
* **Linear:** Output layer for regression tasks

---

### **8) Given a specific problem (e.g., image classification, regression), recommend an activation function and justify your choice.**

* **Problem:** Image Classification
  **Recommended Activation:** ReLU (for hidden layers), Softmax (for output layer)
  **Justification:** ReLU accelerates training and helps avoid vanishing gradients. Softmax gives class probabilities for multi-class outputs.

* **Problem:** Regression
  **Recommended Activation:** ReLU or Linear (output layer)
  **Justification:** Regression requires continuous output, so using linear ensures outputs arent bounded.

---

Let me know if you need diagrams or a summary sheet of activation functions!
 '''
```
